```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://editorial.rottentomatoes.com/guide/best-horror-movies-of-all-time/"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content,"lxml")
```

```python
all_movies = soup.find_all('div' , {"class" : 'col-sm-18 col-full-xs countdown-item-content'})
all_movies[0]
```

```python
movie_names = [movie.find("h2") for movie in all_movies]
movie_names[0]
```

```python
titles = [movie.find("a") for movie in all_movies]
titles = [title.text for title in titles]  #  list of Title names
```

```python
years = [movie.find("span",class_ = "subtle start-year") for movie in all_movies ]
years = [year.string[1:-1] for year in years]
years = [int(year) for year in years]  # List of years;
```

```python
ratings = [movie.find("span",class_="tMeterScore") for movie in all_movies]
ratings = [rating.string.strip("%") for rating in ratings]
ratings = [int(rating) for rating in ratings]   # List of Ratings;
```

```python
# Directors List
directors = soup.find_all("div",class_="info director")
directors[0]
```

```python
dire_names = [dir.find("a") for dir in directors]
dire_names = [name.string for name in dire_names]  # Director names List;
```

```python
# Cast
cast = soup.find_all("div",class_="info cast")
cast[0]
```

```python
cast_movie = [ca.find_all("a") for ca in cast]
cast_all = []
for i in range(len(cast_movie)) :
  l = list(cast_movie[i])
  ca = []
  for j in l :
    ca.append(j.string)
  cast_all.append(ca)
```

```python
cast_all # list of cast of all movies respectively;
```

```python
# Now we have :-
# Titles
# Years
# Ratings
# Directors
# Cast

# we can make a data set with this data;
```

```python
import pandas as pd
movieData = pd.DataFrame()
```

```python
movieData["Title"] = titles
movieData["Director"] = dire_names
movieData["Cast"] = cast_all
movieData["Year"] = years
movieData["Rating(%)"] = ratings
```

```python
movieData.head()
```

```python
# saving the dataset as a csv file;

movieData.to_csv("MovieData_WebScrapping.csv")
```

```python
# sorting the data according to Year;
sorted_data_year = movieData.sort_values(by = "Year")
```
