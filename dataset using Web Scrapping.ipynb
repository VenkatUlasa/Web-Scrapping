```python
import requests

url = "https://quotes.toscrape.com/"
```

```python
# requesting content using url ;
response = requests.get(url)
```

```python
# to know request status  [200 => request is succesful]
response.status_code
```

```python
# Now we need to extract data required;
from bs4 import BeautifulSoup  # BeatufulSoup is a class in bs4 Module;

soup = BeautifulSoup(response.content,"lxml")  # soap is an instance of the class BeautifulSoup we created;
```

```python
# Extracting all the Quotes from the website;
# we need to inspect the website to know the details of tag ;
quotes = soup.find_all("span",class_ = "text")  # here every Quote is in a span tag and with having class name "text".
```

```python
quotes
```

```python
#to get only text part ;
quo_s = [q.text for q in quotes]  # .text is used to separate text part;
```

```python
quo_s
```

```python
# if we do not want " " we can remove;
quo_s = [a[1:-1] for a in quo_s]
quo_s   # it is a list of quotes;
```

```python
# now we need to extract authors ;
authors = soup.find_all("small",class_ = "author")  # in HTML tag is small and class name is "author"
```

```python
aut_s = [a.text for a in authors]
aut_s  # list of author names
```

```python
# now we need to extract tags under quotes;

tags = soup.find_all("div",class_ = "tags")
tags
```

```python
tags_list = []
for i in range(len(tags)) :
  each = []
  for j in tags[i].find_all("a","tag") :
    ts = j.text
    each.append(ts)
  tags_list.append(each)
```

```python
tags_list  # total tag list
```

```python
#  Now we have Quotes,Authors and Tags
# we can create a dataset with this data usind pandas;
import pandas as pd

dataSet = pd.DataFrame()  # creating empty data set.
```

```python
# adding data to the data set;
dataSet["Quote"] = quo_s
dataSet["Tags"] = tags_list
dataSet["Author"] = aut_s
```

```python
dataSet
```

```python
# now we can save the dataset as a csv file;

dataSet.to_csv("WebScrappingFile.csv")
```
